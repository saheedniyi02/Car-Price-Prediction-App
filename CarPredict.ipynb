```python
import pandas as pd
import numpy as np
import pickle
```

```python
data_45=pd.read_csv("Car45Scrap.csv")
data_jumia=pd.read_csv("Carpricejumia.csv")
data_mart=pd.read_csv("CarMart_data.csv")
```

```python
data_mart["Make"]=[name_list[1] for name_list in data_mart["Name"].str.split(" ")] 
```

```python
data=pd.concat([data_45,data_jumia,data_mart])
```

```python
data=data[data["Year"]!="Year"]
```

```python
data["Year"]=data["Year"].astype("i")
```

```python
data.drop("Name",axis=1,inplace=True)
```

```python
def clean_str(string):
    
    string=string.replace("was  6000000","").replace("km","").replace(",","").replace("₦","").replace("\n","").replace(",","")
    return string
```

```python
data["Price"]=[clean_str(price) for price in data["Price"]]
```

```python
data["Price"]=[clean_str(price) for price in data["Price"]]
```

```python
data["Price"]=data["Price"].astype("f")
```

```python
data["Location"]=[loc_string[-1] for loc_string in data["Location"].str.split(",")]
```

```python
data["Location"]=[location.replace(" ","") for location in data["Location"]]
```

```python
oyo_state=["Ibadan","Ogbomosho","Basorun"]
```

```python
rivers_state=["Port Harcourt","Eleme","Harcourt","Alakahia"]
```

```python
delta_state=["Asaba","Warri","Jeddo","Utuoku"]
imo_state=["Owerri"]
anambra_state=["Onitsha","Ichi"]
edo_state=["Benin"]
ondo_state=["Akure"]
benue_state=["Makurdi"]
plateau_state=["Jos"]
kaduna_state=["Zaria","Kaduna","A'Giwa"]
abuja_state=["Abuja","Kubwa","Gwarinpa","Garki","Gwagwalada","Apo","Dawaki"]
nasarawa_state=["Minna","Kokoto"]
enugu_state=["Agu","Uwani"]
ekiti_state=["Ekiti","Adeja"]
abia_state=["Abam"]
bayelsa_state=["Abolgbene"]
akwaibom_state=["Eket"]
```

```python
data["Location"]=data["Location"].replace(lagos_state,"Lagos").replace(ogun_state,"Ogun").replace(kwara_state,"kwara").replace(oyo_state,"Oyo").replace(rivers_state,"Rivers").replace(delta_state,"delta").replace(imo_state,"Imo").replace(anambra_state,"Anambra").replace(edo_state,"Edo").replace(ondo_state,"Ondo").replace(benue_state,"Benue").replace(plateau_state,"Plateau").replace(kaduna_state,"Kaduna").replace(abuja_state,"Abuja").replace(nasarawa_state,"Nasarawa").replace("C45",None).replace(enugu_state,"Enugu").replace(akwaibom_state,"Akwa Ibom").replace(ekiti_state,"Ekiti").replace(abia_state,"Abia").replace(bayelsa_state,"Bayelsa").replace('Cross River','Calabar')
```

```python
data["Location"]=data["Location"].replace(bayelsa_state,"Bayelsa").replace('Cross River','Calabar')
```

```python
data["Mileage"].dtype
```

```python
def clean_mileage(mileage):
    try:
        return mileage.replace("km","").replace(",","")
    except:
            return mileage
```

```python
data["Mileage"]=[clean_mileage(mileage) for mileage in data["Mileage"]]
```

```python
data["Mileage"]=data["Mileage"].astype("f")
```

```python
data_mart.drop("Name",axis=1,inplace=True)
```

```python
data=data[data_mart.columns]
```

```python
data["Transmission"]=data["Transmission"].replace("AUTOMATIC","Automatic")
```

```python
data["Fuel Type"]=data["Fuel Type"].replace({"automatic":None,"manual":None,"Fuel":None})
```

```python
data["Condition"]=data["Condition"].replace("red",None).replace("BLACK",None).replace("GREY",None).replace("WHITE",None).replace("Tokunbo","Foreign Used").replace({"Nigerian Used":"Registered"}).replace("Imported","Foreign Used")
```

```python
data["Make"]=data["Make"].replace({"RAV4":"RAV-4","LandRover":"Land rover"})
```

```python
data["Condition"].replace({"Registered":"Nigerian Used"},inplace=True)
```

```python
data["Model"]=data["Model"].replace({"Cr-V":"CR-V","Lincoln":"licoln","Mdx":"MDX","RAV4":"Rav-4","RAV4":"RAV-4"})
```

```python
data["Model"]=data["Model"].str.capitalize()
```

```python
from sklearn.preprocessing import LabelEncoder 
from sklearn.compose import ColumnTransformer
```

```python
data.isna().sum()
```

```python
object_cols=[]
for i in data.columns:
    if data[i].dtype=="object":
        data[i]=data[i].astype("str")
        object_cols.append(i)
```

```python
for i in data.columns:
    if data[i].dtype=="object":
        enc=LabelEncoder()
        data[i]=data[i].astype("str")
        data[i]=enc.fit_transform(data[i])
```

```python
data
```

```python
target=data["Price"]
train=data.drop("Price",axis=1)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler,RobustScaler 
from sklearn.impute import SimpleImputer 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.pipeline import Pipeline 
from sklearn.ensemble import GradientBoostingRegressor
```

```python
X,val,y,y_val=train_test_split(train,target,test_size=0.1,random_state=42)
```

```python
target
```

```python
data.shape
```

```python
pipe=Pipeline([("imp",SimpleImputer(strategy="constant")),("scaler",StandardScaler()),("model",GradientBoostingRegressor(max_depth=5,random_state=0,max_features=5,learning_rate=0.025,verbose=2,n_estimators=1200))])
```

```python
pipe.fit(X,y)
```

```python
pred=pipe.predict(val)
```

```python
pipe.score(X,y)
```

```python
pipe.score(val,y_val)
```

```python
pred_round=[round(val,-3) for val in pred]
```

```python
mean_absolute_error(pred_round,y_val)
```

```python

pickle.dump(pipe, open('model.pkl','wb'))
```

```python
score=pipe.named_steps["model"].feature_importances_
```
